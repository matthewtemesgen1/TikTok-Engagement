```python
def plot_hist(series, bins=30):
    plt.figure(figsize=(6,4))
    sns.histplot(series.dropna(), bins=bins, kde=True)
    plt.title(f"Distribution of {series.name}")
    plt.xlabel(series.name)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```

```python
plot_hist(df["like_count"])
plot_hist(df["comment_count"])
plot_hist(df["share_count"])
plot_hist(df["play_count"])
plot_hist(df["engagement"])
plot_hist(df["engagement_rate"])
plot_hist(df["caption_length"])
plot_hist(df["hashtag_count"])
```

```python
sns.countplot(data=df, x="author_verified")
plt.title("Verified vs Unverified Creators")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="play_count", y="engagement", alpha=0.3)
plt.xscale("log")
plt.yscale("log")
plt.title("Engagement vs Play Count (Log-Scaled)")
plt.xlabel("Play Count (log scale)")
plt.ylabel("Engagement (log scale)")
plt.tight_layout()
plt.show()
```

```python
df["caption_bin"] = pd.cut(df["caption_length"], 
                           bins=[0, 20, 40, 60, 80, 120], 
                           labels=["0-20", "21-40", "41-60", "61-80", "81-120"])

plt.figure(figsize=(8,4))
sns.boxplot(data=df, x="caption_bin", y="engagement")
plt.title("Engagement by Caption Length Range")
plt.xlabel("Caption Length Bins")
plt.ylabel("Engagement")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x="hashtag_count", y="engagement")
plt.title("Engagement by Number of Hashtags")
plt.xlabel("Hashtag Count")
plt.ylabel("Engagement")
plt.tight_layout()
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), cmap="coolwarm", center=0)
plt.title("Correlation Heatmap of Numeric Variables")
plt.tight_layout()
plt.show()
```
